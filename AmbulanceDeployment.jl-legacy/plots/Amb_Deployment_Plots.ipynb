```python
import json
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_ambulance_deployment(data, description):
    for model in sorted(data.keys()):
        plt.figure(figsize=(12,5))

        for amb_setting in sorted(data[model].keys()):
            if(int(amb_setting) < 25):
                continue
            station_id = list(range(len(data[model][amb_setting])))
            plt.plot(station_id, data[model][amb_setting], label= str(model + ': ' + amb_setting))

            plt.xticks(np.arange(min(station_id), max(station_id)+1, 1.0))
        plt.legend(bbox_to_anchor=(1.04,1), loc="upper left")
        plt.xlabel("Station Id")
        plt.ylabel("# of Ambulances Deployed")
        plt.title(model + " (" + description + ")")
        plt.grid()
        plt.show()
```

## Original Deployment

```python
with open('inputs/og_deployment.json') as f:
    original_data = json.load(f)

print(original_data.keys())
#print(original_data["Stochastic"])
```

```python
plot_ambulance_deployment(original_data, "Original")
```

## New Deployment after Update

```python
with open('inputs/amb_deployment.json') as f:
    updated_data = json.load(f)

print(updated_data.keys())
#print(updated_data["Stochastic"])
```

```python
plot_ambulance_deployment(updated_data, "Updated")
```
